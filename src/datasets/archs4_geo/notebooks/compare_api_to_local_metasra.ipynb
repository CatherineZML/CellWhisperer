```python
import anndata as ad

api = ad.read_h5ad('../data/archs4_annotated.h5ad', backed = True)
local = ad.read_h5ad('../data/archs4_annotated_v2.h5ad', backed = True)
```

```python
api.obs
```

```python
local.obs
```

```python
merged = api.obs.merge(
    local.obs,
    left_index = True,
    right_index = True,
    how = 'inner',
    suffixes = ['_api', '_local']
)
merged
```

```python
from thefuzz import fuzz

different = merged.loc[
    merged[['mapped_ontology_ids_api', 'mapped_ontology_ids_local']].apply(
        lambda x: fuzz.token_set_ratio(x.mapped_ontology_ids_api, x.mapped_ontology_ids_local) < 1,
        axis = 1
    ),
    ['raw_biosample_metadata_api', 'mapped_ontology_ids_api', 'mapped_ontology_ids_local']
]
different
```

```python
for _, row in different.iterrows():
    for key in row.keys():
        print(key, row[key])
```
