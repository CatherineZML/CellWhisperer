```python
import pandas as pd
from metadatamapping import metasra

archs4_metadata = pd.read_csv(
    '../metadata/archs4_sample_metadata.tsv.gz',
    sep = '\t',
    compression = 'gzip'
)

biosample_metadata = pd.read_csv(
    '../metadata/biosample_metadata.csv.gz',
    compression = 'gzip'
)
biosample_metadata = biosample_metadata.drop_duplicates()

sample_accessions = pd.read_csv(
    '../metadata/sample_accessions.tsv.gz',
    sep = '\t',
    compression = 'gzip'
)
sample_accessions = sample_accessions.drop_duplicates()

sra_to_biosample = pd.read_csv(
    '../metadata/sra_to_biosample.csv.gz',
    compression = 'gzip'
)
sra_to_biosample = sra_to_biosample.dropna().drop_duplicates()
sra_to_biosample.biosample = sra_to_biosample.biosample.astype(int)
```

```python
from metadatamapping import metadata
archs4_annotated = metadata.merge_to_annotated_metadata_frame(
    biosample_metadata,
    sra_to_biosample,
    sample_accessions,
    archs4_metadata
)
archs4_annotated
```

```python
accession_and_attributes = metasra.make_accession_and_attributes_table(
    archs4_annotated
)
accession_and_attributes
```
